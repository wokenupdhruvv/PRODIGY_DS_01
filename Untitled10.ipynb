# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Read the Excel file into a DataFrame
# Replace 'your_file.xlsx' with the actual file name
df = pd.read_csv('/Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_567331.csv')

# Step 3: Choose the variable you want to visualize
# For example, 'Age' or 'Gender'
variable = 'Age'  # or 'Gender'

# Step 4: Create a bar chart or histogram
plt.figure(figsize=(10, 6))

# If the variable is categorical (e.g., 'Gender'), use a bar chart
if df[variable].dtype == 'object':
    sns.countplot(x=variable, data=df)
    plt.title(f'Distribution of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Count')

# If the variable is continuous (e.g., 'Age'), use a histogram
else:
    sns.histplot(df[variable], bins=10, kde=True)
    plt.title(f'Distribution of {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frequency')

# Step 5: Show the plot
plt.show()

```python
df = pd.read_csv('/world_population.csv')
```

```python
top_10_pop = df.nlargest(10, '2022 Population')
plt.figure(figsize=(10, 6))
sns.barplot(x='2022 Population', y='Country/Territory', data=top_10_pop)
plt.title('Top 10 Countries by Population in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(df.groupby('Continent')['2022 Population'].sum(), labels=df['Continent'].unique(), autopct='%1.1f%%')
plt.title('World Population Percentage by Continent in 2022')
plt.show()
```

```python
country = 'United States'  # Change this to the country you want to visualize
country_data = df[df['Country/Territory'] == country]
years = [str(year) for year in range(1970, 2023, 5)]
population_values = country_data.iloc[:, 7:12].values.flatten()
plt.figure(figsize=(10, 6))
plt.plot(years[:len(population_values)], population_values, marker='o')
plt.title(f'Population Growth in {country} over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Area (km²)', y='2022 Population', data=df)
plt.title('Population vs Area for Countries in 2022')
plt.xlabel('Area (km²)')
plt.ylabel('Population')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(y='Density (per km²)', data=df)
plt.title('Population Density Distribution in 2022')
plt.ylabel('Density (per km²)')
plt.show()
```
